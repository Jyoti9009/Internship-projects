```python
designation = driver.find_element(By.CLASS_NAME,"suggestor-input  ")
designation.send_keys('Data Scientist')
```

```python
title_tags=driver.find_elements(By.XPATH,'//div [@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in title_tags[:10]:
    title=i.text
    job_title.append(title)
    
location_tags=driver.find_elements(By.XPATH,'//span [@class="locWdth"]')
for i in location_tags[:10]:
    location=i.text
    job_location.append(location)
    
company_tags=driver.find_elements(By.XPATH,'//div [@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div[2]/span/a[1]')
for i in company_tags[:10]:
    company=i.text
    company_name.append(company)
    
experience_tags=driver.find_elements(By.XPATH,'//span [@class="expwdth"]')
for i in experience_tags[:10]:
    experience=i.text
    experience_required.append(experience)
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.azquotes.com/ ")
```

```python
Quote=[]
Author=[]
Process_quote=[]
```

```python
quote_tags=driver.find_elements(By.XPATH, '//div [@class="wrap-block"]/p/a[2]')
for i in quote_tags:
    quote=i.text
    Quote.append(quote)

author_tags= driver.find_elements(By.XPATH, '//div [@class="author"]')
for i in author_tags:
    author=i.text
    Author.append(author)
    
quotes_tags=driver.find_elements(By.XPATH, '//div [@class="tags"]')
for i in quotes_tags:
    quote_text = i.text
    Process_quote.append(quote_text)
```

```python
print(len(Quote),len(Author),len(Process_quote))
```

```python
import pandas as pd
df=pd.DataFrame({'Quote':Quote,'Auhtors':Author,'Type_of_quotes':Process_quote})
df
```

```python
Quotes=[]
Authors=[]
Process_Quotes=[]
```

```python
start=0
end=10
for page in range (start,end):
    quote_tags=driver.find_elements(By.XPATH, '//div [@class="wrap-block"]/p/a[2]')
    for i in quote_tags:
        Quotes.append(i.text)
    next_button=driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]")   
    next_button.click()
    time.sleep(3)
    
start=0
end=10
for page in range (start,end):
    author_tags= driver.find_elements(By.XPATH, '//div [@class="author"]')
    for i in author_tags:
        Authors.append(i.text)
    next_button=driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]")   
    next_button.click()
    time.sleep(3)
    
start=0
end=10
for page in range (start,end):
    quotes_tags=driver.find_elements(By.XPATH, '//div [@class="tags"]')
    for i in quotes_tags:
        Process_Quotes.append(i.text)
    next_button=driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]")   
    next_button.click()
    time.sleep(3)
```
