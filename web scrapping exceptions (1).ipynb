Question 2) In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
driver=webdriver.Chrome()
```

```python
driver.get("http://www.amazon.in/")
```

```python
product = driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
product.send_keys('Guitars')
```

```python
product.send_keys(Keys.RETURN)
```

```python
product_urls=[]
start=0
end=3
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a [@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        product_urls.append(i.get_attribute("href"))
    next_button=driver.find_element(By.XPATH,'//a [@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_button.click()
    time.sleep(2)
```

```python
len(product_urls)  
```

```python
product_urls
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
Brand_names=[]
```

```python
for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    try:
        brand_elements= driver.find_elements(By.XPATH,'//span[@class="a-size-base po-break-word"]')
        if brand_elements:
            Brand_names.append(brand_elements[0].text)
        else:
            Brand_names.append('-')
    except NoSuchElementException:
        Brand_names.append('-')     
```

Question 5) Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.  

```python
driver=webdriver.Chrome()
```

```python
driver.get("http://www.googlemaps.com/")
```

```python
search=driver.find_element(By.XPATH,"/html/body/div[1]/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/form/input")
search.send_keys('Jaipur')
```

```python
search.send_keys(Keys.RETURN)
```

```python
import re
```

```python
current_url = driver.current_url

match = re.search(r'@(-?\d+\.\d+),(-?\d+\.\d+)', current_url)
if match:
    latitude, longitude = match.groups()
    print(f"Latitude: {latitude}, Longitude: {longitude}")
else:
    print("Coordinates not found")
```
